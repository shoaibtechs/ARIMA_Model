# Importing the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from darts import TimeSeries
from darts.models import AutoARIMA
from darts.metrics import mape
```

# Load the Data

```python
# Load your data (assuming it's in a CSV file)
data = pd.read_csv('datasets/AirPassengers.csv')

# Convert the DataFrame into a Darts TimeSeries object
series = TimeSeries.from_dataframe(data, 'Month', '#Passengers')
```

# Split  the Data into training and Validation

```python
train, val = series[:-36], series[-36:]  
```

# Train the ARIMA model

```python
model = AutoARIMA()
model.fit(train)
```

# Make predictions/ Forecasting

```python
forecast = model.predict(len(val))
```

# Evaluating  the model

```python
error = mape(val, forecast)
print(f'MAPE: {error:.2f}%')  #mape mean asbolute percentage error
```

# Visualize Results using Matplotlib

```python
plt.figure(figsize=(10, 6))
series.plot(label='Actual')
train.plot(label='Training')
forecast.plot(label='Forecast')
plt.legend()
plt.show()
```

# Visualize Results using Plotty

```python
train_df = train.pd_dataframe()
val_df = val.pd_dataframe()
forecast_df = forecast.pd_dataframe()

# Create the plot
fig = go.Figure()

# Add actual time series
fig.add_trace(go.Scatter(x=series.time_index, y=series.values(), mode='lines', name='Actual'))

# Add training data
fig.add_trace(go.Scatter(x=train_df.index, y=train_df['#Passengers'], mode='lines', name='Training'))

# Add forecasted data
fig.add_trace(go.Scatter(x=forecast_df.index, y=forecast_df['#Passengers'], mode='lines', name='Forecast'))

# Update layout
fig.update_layout(title='Time Series Forecasting',
                  xaxis_title='Date',
                  yaxis_title='Number of Passengers',
                  legend=dict(x=0, y=1, traceorder='normal'))

# Show plot
fig.show()
```
